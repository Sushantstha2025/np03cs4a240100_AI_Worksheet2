```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/datasets/Copy of medical_students_dataset.csv")
df = pd.DataFrame(data)

df.info()

df_null = df.columns[df.isnull().any()]

print("null value list: ", list[df_null])

df["Student ID"].fillna("Unknown", inplace=True)   # student id must be unique, hence given unknown
df["Age"].fillna(df["Age"].median(), inplace=True)  # age is median as age may contain outliers that may make the central tendency skewed
df["Gender"].fillna("Unknown", inplace=True)  # some people might not be comfortable sharing their gender, hence written unknown
df["Height"].fillna(df["Height"].mean(), inplace=True)  # mean chosen because less chance of outliers, hence normal skewness
df["Weight"].fillna(df["Weight"].mode()[0], inplace=True)  # mode chosen because it represents many people of the same weight. [0] given as mode can be of many values, so take the first mode.
df["Blood Type"].fillna(df["Blood Type"].mode()[0], inplace=True) # mode used to fill because max people have the same blood type.
df["BMI"].fillna(df["BMI"].median(), inplace=True)  # meadian chosen because there might contain outliers
df["Temperature"].fillna(df["Temperature"].median(), inplace=True) # median chosen because some may be outliers
df["Heart Rate"].fillna(df["Heart Rate"].mean(), inplace=True)  #mean chosen because heart rate of every person is average at most times
df["Cholesterol"].fillna(df["Cholesterol"].mode()[0], inplace=True) # mode chosen because taking the most common approach
df["Diabetes"].fillna(df["Diabetes"].mode()[0], inplace=True) # median chosen because some people might be outliers
df["Smoking"].fillna(df["Smoking"].mode()[0], inplace=True) # mode chosen because people either smoke or not and there is no such thing called average/half smoking

print(df.drop_duplicates())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/datasets/Copy of Titanic-Dataset.csv")
df = pd.DataFrame(data)
selected_df = df.loc[df["Pclass"]==1, ["Name", "Pclass", "Sex", "Age", "Fare", "Survived"]]
fare_col = selected_df.loc[:, "Fare"]  # df.loc takes two arguments, row and column, row is given from first to last of Fare column
fare_col.describe()

null_df = df.loc[:, "Age"]
null_df.isna().sum()   # checking the number of null values
null_df = null_df[null_df.notna()]  # keep only rows that are not, null, since age is just a col, it is considered series. that is why, null_df.notna() rather than null_df[null_df["age"]]

embarked_dummies = pd.get_dummies(df["Embarked"], prefix="Embarked")
df = pd.concat([df, embarked_dummies], axis= 1)
df.drop("Embarked", axis= 1, inplace=True)
print(df.head())

df_sex = df.groupby("Sex")
df_survived = df_sex["Survived"]
df_mean = df_survived.mean()

labels = df_mean.index
values = df_mean.values

plt.bar(labels, values, color=['lightblue', 'pink'])
plt.title("Mean Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Sex")
plt.ylim(0, 1) # survival rate ranges from 0 to 1

plt.show()
```
